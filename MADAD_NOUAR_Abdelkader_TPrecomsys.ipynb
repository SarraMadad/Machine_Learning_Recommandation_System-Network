# Part 2: Building a simple recommender

```python
data = pd.read_csv("/home/manelle/movies_metadata.csv")
data.info()
```

#### 1. Choose a metric (or score) to rate the movies on.

```python
m = data["vote_count"].quantile(0.80) # On prends les 20% les mieux classés
m
```

Dans un premier temps, on va sélectionner les films qui sont classés dans les 20 % les mieux notés. On peut voir que 20 % des films ont reçu 50 ou plus de votes.

```python
# The movie on the chart.
# Durée supérieure à 45 minutes et moins de 300 minutes.
temp = data[(data["runtime"] >= 45) & (data["runtime"] <= 300)]
temp.shape
```

Une autre condition préalable que nous voulons mettre en place est la durée d'exécution. Nous ne considérons que les films d'une durée supérieure à 45 minutes
et moins de 300 minutes.

#### 2. Decide on the prerequisites for the movie to be featured on the chart.

```python
chart = temp[temp["vote_count"] >= m] # On garde les meilleurs scores.
chart.head(3)
```

#### 3. Calculate the score for every movie that satisfies the conditions.

```python
# Calcul correspondant au Mean Rating.
C = data['vote_average'].sum()/45466 # sur le nombre d'observations
C
```

```python
v = chart['vote_count']
v
```

```python
R = chart['vote_average']
R
```

```python
# Calcul du Weighted Rating
WR = ((v / (v + m)) * R) + ((m /(v + m)) * C)
WR
```

#### 4. Output the list of movies in decreasing order of their scores.

```python
Score = pd.DataFrame(WR,columns=["Score"])
Score # demander si c'est normal d'en avoir autant 
```

```python
Score['Score'].describe()
```

```python
df = pd.DataFrame(data,columns=['title', 'vote_count', 'vote_average'])
df2 = Score.join(df) 
df2.sort_values(by=['Score'], ascending=False,inplace=True)
df2
```
